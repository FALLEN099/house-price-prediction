# OutLier

```python
#Biểu đồ boxplot
for i, column in enumerate(df_new.columns):
    plt.figure(figsize=(8, 6))
    df.boxplot(column=[column])
    plt.title(f'Boxplot of {column}')
    plt.ylabel('Values')
    plt.show()
```

xử lý Outlier

```python
for column in df_new.columns:    # Tính toán phân vị thứ nhất (Q1) và phân vị thứ ba (Q3)
    Q1 = df_new[column].quantile(0.25)
    Q3 = df_new[column].quantile(0.75)

    # Tính toán phạm vi tứ phân vị (IQR)
    IQR = Q3 - Q1
    # Xác định ngưỡng dưới và trên để xác định outliers
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Lọc ra các giá trị nằm trong ngưỡng
    df_new = df_new[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    # Tạo boxplot cho dữ liệu đã lọc
    plt.figure(figsize=(8, 6))
    sns.boxplot(y=df_new[column], orient='v', width=0.5)
    plt.title(f'Biểu đồ Boxplot của {column} sau khi loại bỏ outliers')
    plt.ylabel('Giá trị')
    plt.show()
```

LinearRegression

```python
# Chọn tất cả các biến đầu vào ngoại trừ 'SalePrice'
X = df_new.drop('SalePrice', axis=1)
y = df_new['SalePrice']

# Chia dữ liệu thành tập huấn luyện và tập kiểm tra theo tỷ lệ 60-40
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Xây dựng mô hình hồi quy tuyến tính
model = LinearRegression()
model.fit(X_train, y_train)

# Dự đoán giá trị trên tập kiểm tra
y_pred = model.predict(X_test)

# Đánh giá hiệu suất của mô hình bằng R^2 Score
r2 = r2_score(y_test, y_pred)

# In ra chỉ số đánh giá (R^2 Score)
print("R^2 Score:", r2)
```

```python
#Biểu đồ đánh giá hiệu suất mô hình
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual SalePrice')
plt.ylabel('Predicted SalePrice')
plt.title('Actual vs. Predicted SalePrice')
plt.show()
```
